## 5.1 (к лекции "Функции")

Исходные данные имеют следующую структуру:

перечень всех документов

documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").


### Задание 1
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

Примеры работы:

1.

Введите команду:
p

Введите номер документа:
10006
Результат:
Владелец документа: Аристарх Павлов

2.

Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Создаем 5 функций выполняющих локальную работу и одну основную (main), внутри которой создаем цикл while, который будет вызывать эти функции

# данная функция по номеру документа находит имя человека в documents

def num_find_name(persons):
    input_number = input('input number ')
    for man in persons:
        if man['number'] == input_number:
            print(man['name'])
        
            
                     # return - что надо вернуть??? Если надо
#num_find_name(documents)

# данная функция по номеру документа находит полку в directories
def num_find_shelf(place):
    input_number = input('input number ')
    for shelf, value in place.items():
        if input_number in value:
            x = shelf
            return x
            
#num_find_shelf(directories)

# выводим все данные по людям и полкам
# непонимаю, как прекратить вывод None. 
def all_info(persons, place):
    for k, v in place.items():
        for man in persons:
            if man['number'] in v:
                print(f"№: {man['number']}, тип: {man['type']}, владелец: {man['name']}, полка хранения: {k}")  
                

#print(all_info(documents, directories))

# создаем новую полку, при условии ее отсутствия
def new_shelf(place):
    input_number = input('Создаем новую полку. input shelf ')
    if input_number in place:
        x = ", ".join(place.keys())
        text = f"Такая полка уже существует. Текущий перечень полок: {x}"
        print(text)
    else:
        place[input_number] = []
        x = ", ".join(place.keys())
        text = f"Полка добавлена. Текущий перечень полок: {x}"
        print(text)
#new_shelf(directories)

# удаляем полку по команде "ds", можно удалить существующую полку из данных только если она пустая
# здесь end отвечает за отсутствие переноса строки

def delete_shelf(place):
    number = input('Удаление полки. Введите номер полки: ')
    if number in list(place.keys()) and place[number] == []:
        del place[number]
        print('Полка удалена. Текущий перечень полок: ', end='')
        print(*list(place.keys()), sep=', ')
    elif number in place and place[number] != []:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', end='')
        print(*list(place.keys()), sep=', ')
    else:
        print('Такой полки не существует. Текущий перечень полок: ', end='')
        print(*list(place.keys()), sep=', ')
        
def main(a, b):
    while True:
        user_input = input('Введите команду (p, s, l, ads, ds, q): ')
        if user_input == 'p':
            num_find_name(documents) #print
        elif user_input == 's':
            print(num_find_shelf(directories))
        elif user_input == 'l':
            all_info(documents, directories) #print
        elif user_input == 'ads':
            new_shelf(directories) #print
        elif user_input == 'ds':
            delete_shelf(directories)
        elif user_input == 'q':
            break
        

main(documents, directories)
```
