### Задание 1
Пункт 1. Пользователь по команде "p" может узнать владельца документа по его номеру

Примеры работы:

1.

Введите команду:
p

Введите номер документа:
10006
Результат:
Владелец документа: Аристарх Павлов

2.

Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Создаем 5 функций выполняющих локальную работу и одну основную (main), внутри которой создаем цикл while, который будет вызывать эти функции

# данная функция по номеру документа находит имя человека в documents

def num_find_name(persons):
    input_number = input('input number ')
    for man in persons:
        if man['number'] == input_number:
            print(man['name'])
                     # return - что надо вернуть??? Если надо
#num_find_name(documents)

# данная функция по номеру документа находит полку в directories
def num_find_shelf(place):
    input_number = input('input number ')
    for shelf, value in place.items():
        if input_number in value:
            print(shelf)
            
#num_find_shelf(directories)

# выводим все данные по людям и полкам
def all_info(persons, place):
    for man in persons:
        text = f"№: {man['number']}, тип: {man['type']}, владелец: {man['name']}"
        #print(text)
        for shelf, value in place.items():
            x = man['number']
            if x in value:
                text2 = f"{text}, полка хранения: {shelf} "
                print(text2) #print(text2)

#print(all_info(documents, directories))

# создаем новую полку, при условии ее отсутствия
def new_shelf(place):
    input_number = input('input shelf ')
    if input_number in place:
        x = list(place.keys())
        text = f"Такая полка уже существует. Текущий перечень полок: {x}"
        print(text)
    else:
        place[input_number] = []
        x = list(place.keys())
        text = f"Полка добавлена. Текущий перечень полок: {x}"
        print(text)
#new_shelf(directories)

# удаляем полку по команде "ds", можно удалить существующую полку из данных только если она пустая



def main(a, b):
    while True:
        user_input = input('Введите команду (p, s, l, ads, ds, q): ')
        if user_input == 'p':
            print(num_find_name(documents))
        elif user_input == 's':
            print(num_find_shelf(directories))
        elif user_input == 'l':
            print(all_info(documents, directories))
        elif user_input == 'ads':
            print(new_shelf(directories))
        elif user_input == 'q':
            break
        

main(documents, directories)
```
